```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('cleaned_Titanic.csv')
df
```

```python
df.isnull().sum()
```

**X & y**

```python
X=df.drop('Survived',axis=1)
y=df['Survived']
```

**Scaling**

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X.iloc[:,[1,2]]=ss.fit_transform(X.iloc[:,[1,2]])
```

```python
X
```

**PCA**

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=0.95)
X=pca.fit_transform(X)
X = pd.DataFrame(X)
```

```python
X.shape
```

**Modelling + Evaluation**

```python
#modelling
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(random_state=True)
model.fit(X,y)

#prediction
ypred=model.predict(X)

#Evaluation
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y,ypred)
#accuracy_test=accuracy_score(y_test,ypred_test)

from sklearn.model_selection import cross_val_score
CV=cross_val_score(model,X,y,cv=5,scoring='accuracy').mean()

print('accuracy:',accuracy)
print('CV score:', CV)
#print('Test accuracy:', accuracy_test)

if(abs(accuracy-CV)<=0.05):
    print('Good Model')
else:
    print('Bad Model')
```
