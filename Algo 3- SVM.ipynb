```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('cleaned_Titanic.csv')
df
```

**X & y**

```python
X=df.drop('Survived',axis=1)
y=df['Survived']
```

**Scaling**

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X.iloc[:,[1,2]]=ss.fit_transform(X.iloc[:,[1,2]])
```

**PCA**

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=0.95)
X=pca.fit_transform(X)
```

```python
X.shape
```

**Hyperparameter tuning**

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

estimator = SVC(random_state = True)
param_grid = {'C' : [0.001, 0.01, 0.1, 1, 10, 100], 'kernel' : ['linear', 'rbf', 'sigmoid', 'poly']}

gs = GridSearchCV(estimator, param_grid, cv = 5, scoring = 'accuracy')
gs.fit(X, y)
gs.best_estimator_
```

**Modelling + Evaluation**

```python
#modeling
model=gs.best_estimator_

#prediction
ypred=model.predict(X)

#Evaluation 
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y,ypred)

from sklearn.model_selection import cross_val_score
CV=cross_val_score(model,X,y,cv=5,scoring='accuracy').mean()

print('accuracy:',accuracy)
print('CV score:', CV)

if (abs(accuracy - CV)<=0.05):
    print('Good Model')
else:
    print('Bad Model')
```

**Save the best model**

```python
from joblib import dump
dump(model,'best_model_SVM.joblib')
```
