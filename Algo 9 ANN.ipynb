**X &y**

```python
X=df.drop(columns='Survived')
y=df['Survived']
```

**Initializing the artificial neural network**

```python
import keras
import tensorflow as tf
```

```python
from keras.models import Sequential
from keras.layers import Dense
model=Sequential()
model.add(Dense(input_dim=30,units=60,kernel_initializer='uniform',activation='relu'))
model.add(Dense(units=1,kernel_initializer='uniform',activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(X,y,batch_size=32,epochs=100)
```

```python
model.summary()
```

```python
y_pred=model.predict(X)>=0.5

# Evaluation
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y,y_pred)
print('Accuracy:',accuracy)
```

**Cross validate the model**

```python
def build_cross_classifier():
    classifier=Sequential()
    classifier.add(Dense(input_dim=30,units=60,kernel_initializer='uniform',activation='relu'))
    classifier.add(Dense(units=1,kernel_initializer='uniform',activation='sigmoid'))

    classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
    return classifier
    
```

```python
from scikeras.wrappers import KerasClassifier
classifier=KerasClassifier(build_cross_classifier,batch_size=32,epochs=100)
```

```python
from sklearn.model_selection import cross_val_score
CV=cross_val_score(classifier,X,y,cv=5,scoring='accuracy').mean()
```

```python
print('CV:',CV)
```
